```python
from sklearn.model_selection import train_test_split

# Labels sont les valeurs que nous souhaitons prédire 
labels = np.array(df['Rating'])

# 30% examples in test data
train, test, train_labels, test_labels = train_test_split(df, labels, 
                                                          test_size = 0.2, 
                                                          random_state = 42)
```

```python
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
```

```python
def convert(data):
    number = preprocessing.LabelEncoder()
    data['GENDER'] = number.fit_transform(data.GENDER)
    data['WORKING'] = number.fit_transform(data.WORKING)
    data['REGION'] = number.fit_transform(data.REGION)
    data['MUSIC'] = number.fit_transform(data.MUSIC)
    data['HEARD_OF'] = number.fit_transform(data.HEARD_OF)
    data['LIST_OWN'] = number.fit_transform(data.LIST_OWN)
    data['LIST_BACK'] = number.fit_transform(data.LIST_BACK)
    data['OWN_ARTIST_MUSIC'] = number.fit_transform(data.OWN_ARTIST_MUSIC)
    return data
```

```python
convert(train)
```

```python
convert(test)
```

```python
model.fit(train, train_labels)
```

```python
# Make probability predictions
train_probs = model.predict_proba(train)[:, 1]
probs = model.predict_proba(test)[:, 1]

train_predictions = model.predict(train)
predictions = model.predict(test)
```

```python
from sklearn.metrics import precision_score, recall_score, roc_auc_score, roc_curve

print(f'Train ROC AUC Score: {roc_auc_score(train_labels, train_probs)}')
print(f'Test ROC AUC  Score: {roc_auc_score(test_labels, probs)}')
```
